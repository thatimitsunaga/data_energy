```python
# Libraries
import pandas as pd
import datetime
```

```python
# alocando os dados do github da tb_canais
url_canais = "https://raw.githubusercontent.com/thatimitsunaga/data_energy/main/tb_canais.csv"
```

```python
# criando variavel para dados da tb_canais
canais = pd.read_csv(url_canais, encoding="latin1", sep=";")
canais.head()
display(canais)
```

```python
print(canais["COD_INSTALACAO"].value_counts())
```

```python
# mesma etapa para tb_servicos
url_servicos = "https://raw.githubusercontent.com/thatimitsunaga/data_energy/main/tb_servicos.csv"
```

```python
servicos = pd.read_csv(url_servicos, encoding="latin1", sep=";")
servicos.head()
display(servicos)
```

```python
print(servicos["COD_INSTALACAO"].value_counts())
```

```python
print(servicos["DT_CRIACAO"].dtype)
```

```python
# converter coluna dt_criacao para datetime
servicos["DT_CRIACAO"] = pd.to_datetime(servicos["DT_CRIACAO"], dayfirst=True)

servicos["DT_CRIACAO"].dt.month.value_counts()
```

```python
display(servicos)
```

```python
filter_july = servicos[
    (servicos["DT_CRIACAO"].dt.month == 7)
]
```

```python
display(filter_july)
```

```python
# filtrando ocorrencias de religacao em julho
servicos_julho_religacao = servicos[
    (servicos["SERVICO"].str.contains("Religação")) &
    (servicos["DT_CRIACAO"].dt.month == 7)
]
```

```python
display(servicos_julho_religacao)
```

```python
print(servicos_julho_religacao["SERVICO"].unique())
```

```python
# juntando os dados de religacao de julho com as demais infos (canal, no caso)

dados_completos = pd.merge(servicos_julho_religacao, canais, on="COD_INSTALACAO", how="inner")

display(dados_completos)
```

```python
print(dados_completos["COD_INSTALACAO"].value_counts())
```

```python
# checando mais obs/linhas depois do merge
duplicados = dados_completos[dados_completos.duplicated(subset='COD_INSTALACAO', keep=False)]
print(duplicados)
```
